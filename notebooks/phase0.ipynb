## Load data

```python
import pandas as pd

eod_data = pd.read_parquet('/home/noahg/MATH316/project2/data/0004691-251025141854904.parquet')
print(len(eod_data))
eod_data['eventDate'] = pd.to_datetime(eod_data['eventDate'], errors='coerce')
eod_data.head(n=5)
```

## Get counts data

Each row in the dataset is the results of a survey by someone. The individualCount column is how many of that species of bird the person observed. This field can be NaN though as not all people track counts during their surveys. Because of this, we split the dataset so one subset contains only records where the count was tracked by the surveyor.

```python
eod_counts = eod_data.loc[eod_data['individualCount'].notna()]
eod_counts['eventYear'] = eod_counts['eventDate'].dt.year
eod_counts.head(n=5)
```

## Build hierarchical tables


```python

ht1 = eod_counts.groupby(['countryCode','eventYear','species'])['individualCount'].sum().reset_index().set_index(['countryCode','eventYear','species']).sort_values(by='individualCount', ascending=False)
ht1.head(n=20)
```
